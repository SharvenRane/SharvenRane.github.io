### What is Statistics?
Statistics is the science of collecting, organizing and analyzing data.

### What  is Descriptive Stats?
It consists of organzing and summarizing data

### What is Inferential Stats?
Techniques used on data that we measure to form conclusions

### Population (N)
A population refers to the entire group of individuals, items, or events
### Sample (n)
A sample is a subset of the population that is selected for analysis.
### Sampling Techniques
#### Probability Sampling:
1. Simple Random Sampling: Every member of the population has an equal chance of selection. Eg. All humans
   
2. Stratified Sampling: Population is divided into non over lapping subgroups, and samples are taken proportionally from each. Eg. Male Female , (0-10 age) (20-50).

3. Systematic Sampling: Select every n th member after a random starting point from the population.Eg. Every 8th person, or every 7th person

4. Cluster Sampling: Randomly select entire clusters and include all members within them. Eg. Cluster of age or gender or salary.

5. Multi-Stage Sampling: Combine multiple sampling techniques, such as clusters and random selection. Eg. From a country 10 states from those 10 states 25 universities from those 25 universities 10000 students.
#### Non-Probability Sampling:
1. Convenience Sampling: Select the easiest members to access based on maybe surveys. Eg. People who know data science. or who are useful only to your particular research or analysis.
   
2. Judgmental (Purposive) Sampling: Select members based on the researcher’s judgment. eg. a researcher studying the buying patterns of high-end luxury car owners

3. Quota Sampling: Ensure specific quotas for subgroups in the sample. eg a market researcher conducting a survey about a new product

4. Snowball Sampling: Existing participants recruit others from their network. eg high paying test survey people will tell more people to earn the money for participation

## 1. Descriptive Statistics

### Mean: The average of a set of values.
### Formula: mean = sum(Of all the 1000 values) / Total number of values (here its 1000)

```python
mean_value = np.mean(data)
mean_value
```

### Median: The middle value of a dataset when ordered from least to greatest.
### Formula: Median = {(n ÷ 2)th value + [(n ÷ 2) + 1]th value} ÷ 2  FOR EVEN & Median = (n + 1) ÷ 2 FOR ODD
### Here n is number of data points which is 1000

```python
median_value = np.median(data)
median_value
```

### Mode: The value that appears most frequently in a dataset. 
### Note: We use stats.mode to calculate mode.

```python
# Calculate the mode
mode_value = stats.mode(data)
mode_value
```

### Standard Deviation: A measure of the spread or dispersion of a dataset.
### A small std means the numbers are close to each other (like 0 to 3) and the mean. 
### A big std means the numbers are spread out from the mean (like 10, 50, or higher).
### Formula: std = calculates the squared difference between each data point and the mean, then averages those squared differences (divided by N for population or n−1 for sample) and takes the square root to give the standard deviation.

```python
std_dev = np.std(data)
std_dev
```


### Variance: A measure of how much the numbers in a dataset differ from the mean.
### A small variance means the numbers are close to each other (like 0 to 3) and the mean.
### A big variance means the numbers are spread out from the mean (like 10, 50, or higher).
### Formula: Variance = calculates the squared differences between each data point and the mean, averages those squared differences (divided by N for population or n−1 for sample), and gives a value that represents the spread of the data. Variance is the step before taking the square root to find the standard deviation.

```python
variance_value = np.var(data)
variance_value
```

### Difference between variance and std
### Variance gives a more mathematical measure of spread but is in squared units.
### Standard deviation provides a more practical measure of spread in the same units as the data.

## C. Inferential Statistics

### Non-Parametric Tests (e.g., Mann-Whitney, Kruskal-Wallis, Friedman) are used when the data does not follow a normal distribution.
### Parametric Tests (e.g., t-tests, ANOVA) assume that the data follows a normal distribution

### 1. One-sample t-test to check if the sample mean is significantly different from a value.
### Null Hypothesis (H0): The sample mean is equal to 0.
### Alternative Hypothesis (H1): The sample mean is not equal to 0.

```python
t_stat, p_value = stats.ttest_1samp(data, 0)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

### A large positive or negative value indicates a large difference between the sample mean and the hypothesized mean.
### A value near 0 means the sample mean is close to the hypothesized mean.

### If p_value < significance level (e.g., 0.05): Reject the null hypothesis
### If p_value >= significance level: Fail to reject the null hypothesis

### 2. Two-Sample t-Test (Independent Samples)

```python
from scipy.stats import ttest_ind

# Generate random data for two groups
group1 = np.random.normal(loc=5, scale=1.5, size=30)  # Generate 30 random values for group1 from a normal distribution with mean=5 and std=1.5
group2 = np.random.normal(loc=6, scale=1.2, size=30)  # Generate 30 random values for group2 from a normal distribution with mean=6 and std=1.2

# Perform two-sample t-test
t_stat, p_value = ttest_ind(group1, group2)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

### 3. Paired t-Test (Related Samples)

```python
from scipy.stats import ttest_rel

# Generate paired data
before = np.random.normal(loc=5, scale=1.5, size=30)
after = before + np.random.normal(loc=0.5, scale=0.5, size=30)

# Perform paired t-test
t_stat, p_value = ttest_rel(before, after)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

### 4. ANOVA (Analysis of Variance)
### Compare means across three or more groups

```python
from scipy.stats import f_oneway

# Generate data for three groups
group1 = np.random.normal(loc=5, scale=1.5, size=30)
group2 = np.random.normal(loc=6, scale=1.5, size=30)
group3 = np.random.normal(loc=7, scale=1.5, size=30)

# Perform ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)
print(f"F-statistic: {f_stat}, P-value: {p_value}")
```

### A large F-statistic means the differences between group means are significant.
### A small F-statistic means the differences between group means are not significant.
### Null Hypothesis (H₀): The average test scores of students in Class A, Class B, and Class C are the same.
### Alternative Hypothesis (H₁): At least one class has a different average test score.

### 5. Chi-Square Test
### To see if there's a significant relationship between the two categorical variables.

```python
from scipy.stats import chi2_contingency

# Let's say we are looking at the relationship between Gender (Male/Female) and Preference (Tea/Coffee)
observed = np.array([[30, 10],  # 30 males like tea, 10 like coffee
                     [20, 40]])  # 20 females like tea, 40 like coffee

chi2, p, dof, expected = stats.chi2_contingency(observed)

print(f"Chi-Square: {chi2}, P-value: {p_value}, Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)
```

### Contingency Table: A table that shows the frequency distribution of variables. Observed data showing how many people fall into each category.
### chi2: The chi-square test statistic, which tells us the difference between observed and expected frequencies.
### Large chi2 value means dependent. Small chi2 value means independent
### dof (degrees of freedom): The number of independent pieces of information used to calculate the statistic.
### expected: The expected frequencies represent what you would expect to observe in each category if gender and preference were independent.
### Null Hypothesis (H₀): There is no association between the two variables (i.e., the variables are independent).
### Alternative Hypothesis (H₁): There is an association between the two variables (i.e., the variables are dependent).

### 6. Correlation Tests (Pearson & Spearman)

### Pearson Correlation
### When both variables are continuous and have a linear relationship.
### When you expect the data to follow a normal distribution.

```python
from scipy.stats import pearsonr

# Sample data
x = [10, 20, 30, 40, 50]
y = [12, 24, 33, 45, 55]

corr, _ = pearsonr(x, y)

print(f"Pearson correlation coefficient: {corr}")
```

### 1 means a perfect positive linear relationship.
### -1 means a perfect negative linear relationship.
### 0 means no linear relationship.

### Spearman's Rank Correlation
### When the relationship between the variables is monotonic, but not necessarily linear.
### When the data is ordinal or not normally distributed.
### Monotonic relationship means that as one variable increases, the other either increases or decreases, but not necessarily at a constant rate

```python
from scipy.stats import spearmanr

# Sample data
x = [10, 20, 30, 40, 50]
y = [12, 24, 33, 45, 55]

corr, _ = spearmanr(x, y)

print(f"Spearman correlation coefficient: {corr}")
```

### 1 means a perfect positive monotonic relationship.
### -1 means a perfect negative monotonic relationship.
### 0 means no monotonic relationship.

### Pearson is great for linear relationships with continuous data.
### Spearman works well when the data doesn’t necessarily follow a linear trend but maintains a monotonic relationship.

### Mann-Whitney U Test (also called the Wilcoxon Rank-Sum Test) (Non parametric)
### When you have two independent samples.
### When the data is ordinal or not normally distributed.
### Null Hypothesis (H₀): The two groups have the same distribution (no significant difference).
### Alternative Hypothesis (H₁): The two groups have different distributions.
### The calculated value of U, tells you the degree of difference between the two groups.
### U = 0: One group has smaller values than the other. Larger U means groups are more similar.

```python
from scipy.stats import mannwhitneyu

# Sample data from two groups
group1 = [12, 15, 18, 22, 25]
group2 = [30, 35, 40, 45, 50]

stat, p_value = mannwhitneyu(group1, group2)

print(f"U-statistic: {stat}, p-value: {p_value}")
```

### Wilcoxon Signed-Rank Test (Non-parametric)
### When you have two related samples or matched pairs.
### When the data is ordinal or not normally distributed.
### Null Hypothesis (H₀): The median difference between the two groups is zero (no significant difference).
### Alternative Hypothesis (H₁): The median difference between the two groups is not zero.
### W-statistic: Represents the rank sum of the differences, which gives insight into the magnitude of changes.
### W = 0: No difference in pairs. Larger W means more positive differences between the pairs.

```python
from scipy.stats import wilcoxon

# Sample data from two related groups
group1 = [12, 15, 18, 22, 25]
group2 = [30, 35, 40, 45, 50]

stat, p_value = wilcoxon(group1, group2)

print(f"Test statistic: {stat}, p-value: {p_value}")
```

### Kruskal-Wallis H Test (Non-parametric)
### When you have three or more independent groups.
### When the data is ordinal or not normally distributed.
### Null Hypothesis (H₀): The distributions of all groups are the same.
### Alternative Hypothesis (H₁): At least one group has a different distribution.
### H-statistic: Reflects the degree to which the ranks are different across the groups. Larger values indicate that the groups differ more from each other.
### H = 0: No difference between groups. Larger H means greater difference between groups.

```python
from scipy.stats import kruskal

# Sample data from three groups
group1 = [12, 15, 18, 22, 25]
group2 = [30, 35, 40, 45, 50]
group3 = [60, 65, 70, 75, 80]

stat, p_value = kruskal(group1, group2, group3)

print(f"H-statistic: {stat}, p-value: {p_value}")
```

### Friedman Test (Non-parametric)
### When you have three or more related groups.
### When the data is ordinal or not normally distributed.
### Null Hypothesis (H₀): The distributions of all groups are the same.
### Alternative Hypothesis (H₁): At least one group has a different distribution.
### Test statistic: Measures difference in ranks across repeated measures
### χ²_F = 0: No difference in ranks. Larger χ²_F means greater difference between conditions.

```python
from scipy.stats import friedmanchisquare

# Sample data from three related groups
group1 = [12, 15, 18, 22, 25]
group2 = [30, 35, 40, 45, 50]
group3 = [60, 65, 70, 75, 80]

# Friedman test
stat, p_value = friedmanchisquare(group1, group2, group3)

print(f"Test statistic: {stat}, p-value: {p_value}")
```

## D. Probability Theory

### 1. Probability is the measure of the likelihood that an event will occur. It ranges from 0 (impossible) to 1 (certain).

```python
# Probability of heads in a fair coin toss
heads_prob = 0.5
tails_prob = 1 - heads_prob
print(f"Probability of heads: {heads_prob}, Probability of tails: {tails_prob}")
```

### 2. Conditional probability is the probability of an event occurring given that another event has already occurred.
### P(A∣B)= P(A∩B) / P(B)

```python
# P(A|B) = P(A and B) / P(B)
P_A_and_B = 0.2  # Probability that A and B both happen
P_B = 0.4        # Probability that B happens
P_A_given_B = P_A_and_B / P_B
print(f"Conditional Probability P(A|B): {P_A_given_B}")
```

### 3. Bayes' Theorem 
### 𝑃(𝐴∣𝐵) indirectly when you have the prior (𝑃(𝐴)) and the likelihood (𝑃(𝐵∣𝐴))

```python
# Bayes' Theorem calculation
P_A = 0.01  # Probability of A happening
P_B_given_A = 0.9  # Probability of B happening given A has already occured
P_B = 0.05  # Probability of B happening

P_A_given_B = (P_B_given_A * P_A) / P_B
print(f"Posterior Probability P(A|B): {P_A_given_B}")
```

### 4. Random Variables:
### A random variable is a variable whose value is subject to randomness. It can be either discrete (like the number of heads in 10 coin flips) or continuous (like the height of a person).

```python
# Discrete random variable: Rolling a die
die_rolls = np.random.randint(1, 7, size=1000)
print(f"Sample of die rolls: {die_rolls[:10]}")
```

### 5. Law of Total Probability:
### It is a rule for finding the probability of an event by considering all possible ways that the event can occur.

```python
P_A_given_B = 0.7  # Passing given studied
P_A_given_C = 0.3  # Passing given didn't study
P_B = 0.6          # Probability of studying
P_C = 1 - P_B      # Probability of not studying

P_A = P_A_given_B * P_B + P_A_given_C * P_C
print(f"Total Probability P(A): {P_A}")
```

### 6. Markov Chains:
### A Markov chain is a mathematical system that undergoes transitions from one state to another, with the probability of each transition depending only on the current state.

```python
states = ['Sunny', 'Rainy']
transition_matrix = [[0.8, 0.2], [0.4, 0.6]]  # Transition probabilities
current_state = 0  # Start with Sunny

next_state = np.random.choice([0, 1], p=transition_matrix[current_state])
print(f"Next state (0: Sunny, 1: Rainy): {states[next_state]}")
```
